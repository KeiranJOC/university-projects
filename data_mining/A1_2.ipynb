```python
from math import sqrt

def pearson_correlation(df, x, y):
    n = len(df.index)
    sum_xy = (df[x] * df[y]).sum()
    sum_x = df[x].sum()
    sum_y = df[y].sum()
    sum_x2 = (df[x]**2).sum()
    sum_y2 = (df[y]**2).sum()

    return (n * sum_xy - sum_x * sum_y) / sqrt((n * sum_x2 - sum_x**2) * (n * sum_y2 - sum_y**2))
```

```python
def euclidean_distance_matrix(dataframe):
    matrix = dataframe.to_numpy()
    euclidean_dm = np.zeros((matrix.shape[0], matrix.shape[0]))
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[0]):
            distance = np.linalg.norm(matrix[i] - matrix[j])
            euclidean_dm[i,j] = distance
            
    return pd.DataFrame(euclidean_dm, index=dataframe.index, columns=dataframe.index)
```

```python
from Graph import *

def gabriel_graph(dataframe):
    gg = Graph(vertices=dataframe.index.to_list())
    matrix = dataframe.to_numpy()

    for i in range(len(matrix)-1):
        for k in range(i+1, len(matrix)):
            d_uv = matrix[i,k]
            valid = True

            for j in range(len(matrix)):
                if not j == i and not j == k:
                    d_uw = matrix[i,j]
                    d_vw = matrix[j,k]
                    if d_uv**2 > d_uw**2 + d_vw**2:
                        valid = False
                        break

            if valid:
                gg.add_edge_directed_by_index(u_index=i, v_index=k, weight=d_uv)
    
    return gg
  
```

```python
def set_cover(universe, subsets):
    elements = set(e for s in subsets for e in s)

    # Check the subsets cover the universe
    if elements != universe:
        return None
    covered = set()
    cover = []

    # Greedily add the subsets with the most uncovered points
    while covered != elements:
        subset = max(subsets, key=lambda s: len(s - covered))
        cover.append(subsets.index(subset))
        covered |= subset
 
    return cover
```

```python
def feature_set(input_matrix, target_vector):
    
    # generate list of different-class pairs
    values, counts = np.unique(target_vector, return_counts=True)
    class_1, class_2 = values[0], values[1]
    class_1_count, class_2_count = counts[0], counts[1]
    x = np.array(list(range(class_1_count)))
    y = np.array(list(range(class_1_count, class_1_count + class_2_count)))
    pairs = np.transpose([np.tile(x, len(y)), np.repeat(y, len(x))])

    # generate matrix of feature difference between different-class pairs
    difference_matrix = np.zeros((pairs.shape[0], input_matrix.shape[1]), dtype=np.int8)

    for i, pair in enumerate(pairs):
        sample_1, sample_2 = input_matrix[pair[0]], input_matrix[pair[1]]
        mask = np.equal(sample_1, sample_2)
        indices = np.nonzero(~mask)
        difference_matrix[i, indices] = 1

    # find infeasible rows that contain only zeros
    infeasible_rows = [] 

    for i, row in enumerate(difference_matrix):
        if not np.any(row):
            infeasible_rows.append(i)

    # remove infeasible rows from consideration
    universe = set(range(difference_matrix.shape[0])) - set(infeasible_rows)
    subsets = []

    # pass columns as subsets to set cover algorithm
    for i in range(difference_matrix.shape[1]):
        column = difference_matrix[:,i]
        subset = np.nonzero(column)[0]
        subsets.append(set(subset))

    cover = set_cover(universe, subsets)

    return sorted(cover)
```

### K-NN classifier

```python
def hamming_distance(sample_1, sample_2):
    return np.bitwise_xor(sample_1, sample_2).sum()


def euclidean_distance(sample_1, sample_2):
    return np.linalg.norm(sample_1 - sample_2)


def get_neighbours(x_train, y_train, test_sample, k, distance_measure):
    distances = []
    for i, training_sample in enumerate(x_train):
        if distance_measure == 'hamming':
            distance = hamming_distance(test_sample, training_sample)
        elif distance_measure == 'euclidean':
            distance = euclidean_distance(test_sample, training_sample)
        target = y_train[i]
        distances.append((training_sample, distance, target))

    distances.sort(key=lambda x: x[1])
    return [distances[i] for i in range(k)]


def predict(x_train, y_train, test_sample, k, distance_measure):
    neighbours = get_neighbours(x_train, y_train, test_sample, k, distance_measure)
    distances = [sample[1] for sample in neighbours]
    targets = [sample[2] for sample in neighbours]   

    freq_1, freq_2 = 0, 0
    for neighbour in neighbours:
        distance = neighbour[1]
        target = neighbour[2]
        if target == 0:
            freq_1 += (1 / distance)
        elif target == 1:
            freq_2 += (1 / distance)

    return 0 if freq_1 > freq_2 else 1


def k_nearest_neighbours(x_train, y_train, x_test, k, distance_measure):
    predictions = [] 
    for sample in x_test:
        prediction = predict(x_train, y_train, sample, k, distance_measure)
        predictions.append(prediction)
    return predictions
```

### Classification metrics

```python
def classification_metrics(actual, predicted):
    tp, tn, fp, fn = 0, 0, 0, 0
    
    for a, p in zip(actual, predicted):
        if a == 0 and p == 0:
            tn += 1
        elif a == 0 and p == 1:
            fp += 1
        elif a == 1 and p == 0:
            fn += 1
        elif a == 1 and p == 1:
            tp += 1

    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    precision = tp / (tp + fp)
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    f1 = 2 / (1 / precision + 1 / sensitivity)
    
    mcc_denominator = sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
    if not mcc_denominator == 0:
        mcc = (tp * tn - fp * fn) / sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
    else:
        mcc = 0
        
    j = sensitivity + specificity - 1

    print(f"Sensitivity:                        {sensitivity}")
    print(f"Specificity:                        {specificity}")
    print(f"Accuracy:                           {accuracy}")
    print(f"F1-score:                           {f1}")
    print(f"Matthew's Correlation Coefficient:  {mcc}")
    print(f"Youden's J Statistic:               {j}")

    number_of_classes = np.unique(actual).shape[0]
    confusion_matrix = np.zeros((number_of_classes, number_of_classes), dtype='int8')
    confusion_matrix[0,0] = tp
    confusion_matrix[0,1] = fp
    confusion_matrix[1,0] = fn
    confusion_matrix[1,1] = tn

    return confusion_matrix
```

### Select features using ANOVA F-values

```python
def anova_f_score(dataframe, feature, target, N, n, df_between, df_within):
    
    sum_squares_between = (sum(dataframe.groupby(target).sum()[feature]**2) / n) - (dataframe[feature].sum() ** 2 / N)
    sum_y_squared = sum([value**2 for value in dataframe[feature].values])
    sum_squares_within = sum_y_squared - sum(dataframe.groupby(target).sum()[feature]**2) / n
    sum_squares_total = sum_y_squared - (iris_dataset_df[feature].sum()**2) / N

    mean_square_between = sum_squares_between / df_between
    mean_square_within = sum_squares_within / df_within

    return mean_square_between / mean_square_within


def anova_feature_selection(dataframe, features, target, k):
    m = len(pd.unique(iris_dataset_df[target]))
    N = iris_dataset_df.shape[0]
    n = iris_dataset_df.groupby(target).size()[0]

    df_between = m - 1
    df_within = N - m
    df_total = N - 1

    f_values = []

    for feature in features:
        f_value = anova_f_score(dataframe, feature, target, N, n, df_between, df_within)
        f_values.append((feature, f_value))

    f_values.sort(key=lambda x: x[1], reverse=True)

    return [f_values[i] for i in range(k)]
```

```python
def hamming_distance_unicode(sample_1, sample_2):
    distance = 0
    indices = []
    for k, (i, j) in enumerate(zip(sample_1, sample_2)):
        if not i == j:
            distance += 1
            indices.append(k)
    return distance, indices


def get_distances(P):
    distances = []
    for i, sample_1 in enumerate(P):
        for j, sample_2 in enumerate(P):
            distance, indices = hamming_distance_unicode(sample_1, sample_2)
            if not i == j:
                distances.append([i, j, distance, indices])

    return sorted(distances, key=lambda x: x[2])
```

```python
def is_string_compatible(p, g):
    for i in range(len(p)):
        if p[i] == '*':
            continue
        if not p[i] == g[i]:
            return False
    return True


def is_G_compatible(P, G):
    G_compatible = True
    
    for i in range(G.shape[0]):
        g = G[i]
        compatible = False
        for j in range(P.shape[0]):
            p = P[j]
            if is_string_compatible(p, g):
                compatible = True
                break
        if not compatible:
            G_compatible = False
            return G_compatible

    return G_compatible


def is_B_compatible(P, B):
    B_compatible = False
    
    for b in B:
        for p in P:
            if is_string_compatible(p, b):
                B_compatible = True
                break

    return B_compatible   
```

```python
def make_compatible(p1, p2, indices):
    p3 = p2.copy()
    for i in indices:
        p3[i] = '*'
    if is_string_compatible(p3, p1):
        return p3
    else:
        return p2


def reduce_pattern_set(P, B, l):   
    while P.shape[0] > l:
        distances = get_distances(P)
        P_copy = P.copy()

        for i in range(len(distances)):
            instance = distances[i]
            p1_index = instance[0]
            p2_index = instance[1]
            indices = instance[-1]
            p1 = P[p1_index]
            p2 = P[p2_index]
            p3 = make_compatible(p1, p2, indices)
            P[p2_index] = p3
            P = remove_redundant_patterns(P)
            
            if not is_B_compatible(P, B):
                break
            else:
                P = P_copy

    return P


def remove_redundant_patterns(P):
    indices = []
    removed = []
    for i in range(P.shape[0]):
        for j in range(P.shape[0]):
            if i == j:
                continue
            if is_string_compatible(P[i], P[j]):
                indices.append(j)

    for i in indices:
        removed.append(list(P[i]))
    P_reduced = np.delete(P, indices, axis=0)

    return P_reduced
```

```python
def evaluate_cost(P, G, B):
    G_compatible = is_G_compatible(P, G)
    B_compatible = is_B_compatible(P, B)
    
    if not G_compatible or B_compatible:
        return 1000
    
    return P.shape[0]


def get_neighbourhood(P):
    indices = np.nonzero(np.logical_not(P == '*'))
    neighbourhood = []

    for k, (i, j) in enumerate(zip(indices[0], indices[1])):
        P_copy = P.copy()
        P_copy[i,j] = '*'
        P_reduced = remove_redundant_patterns(P_copy)
        neighbourhood.append(P_reduced)
    
    return neighbourhood


def choose_neighbour(N):
    rng = np.random.default_rng()
    costs = np.zeros((len(N)), dtype='int8')
   
    for i in range(len(N)):
        n = N[i]
        costs[i] = evaluate_cost(n, G, B)

    indices = np.nonzero(costs == costs.min()) 
    index = rng.choice(len(N))

    return N[index]


def local_search(P, G, B, max_tries):
    P_min = P
    cost = evaluate_cost(P_min, G, B)

    for i in range(max_tries):
        N = get_neighbourhood(P_min)
        n = choose_neighbour(N)
        if evaluate_cost(n, G, B) < evaluate_cost(P_min, G, B):
            P_min = n
            cost = evaluate_cost(P_min, G, B)
            print(f'{i} iterations: current minimum = {cost}')

    return P_min, cost
```
